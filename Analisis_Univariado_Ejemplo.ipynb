En algunos conjuntos de datos, hay valores (puntos de datos observados), llamados valores atípicos. En estadística, tales como muestras estratificadas, un valor atípico (en inglés outlier) es una observación que es numéricamente distante del resto de los datos.

Veamos el impacto de los datos atípicos en forma gráfica.

```python
# Veamos el impacto de los datos atípicos en forma gráfica
fig, axes = plt.subplots(1,2)

sns.boxplot(data=df['SkinThickness'], ax = axes[0]).set_title('Valores Atípicos')
sns.boxplot(data=df['Pregnancies'], ax = axes[1]).set_title('Boxplot Normal')

plt.show()
```

```python
# Muchas veces se integran valores atípicos como una representación
# de un valor faltante o nulo, es importante conocer si en la descripción
# de nuestro dataset se menciona la razón de este tipo de valores.

print('El promedio de variable SkinThickness es:', df['SkinThickness'].mean())
```

¿Qué significa la presencia de un dato atípico en nuestro dataset? ¿Nos impacta en algo los datos atípicos? Veamos el ejemplo anterior.

```python
# Aplicamos limpieza de datos para valores faltantes y valores atípicos

# Creamos una copia del data frame para no perder el original
df_clean = df.copy()

# Eliminamos los datos nulos del data frame
df_clean.dropna(inplace=True) # El argumento inplace=True en df_clean.dropna(inplace=True) tiene como objetivo modificar directamente el DataFrame original en lugar de crear una copia con los cambios realizados.

# Eliminamos valores atípicos del dataframe
df_clean = df_clean[df_clean['SkinThickness'] != 9999]
df_clean = df_clean[df_clean['DiabetesPedigreeFunction'] != -1]
```

```python
# Verificamos que los datos nulos del data frame hayan sido eliminadas
df_clean.isna().sum()
```

```python
# Veamos el resultado de eliminar los datos atípicos
fig, axes = plt.subplots(1,2)

sns.boxplot(data=df_clean['SkinThickness'], ax = axes[0]).set_title('Sin Valores Atípicos')
sns.boxplot(data=df_clean['Pregnancies'], ax = axes[1]).set_title('Boxplot Normal')

plt.show()
```

```python
# Creamos una copia del data frame para no perder el original
df_prueba = df.copy()
# Eliminamos valores atípicos del dataframe
df_prueba = df_prueba[df_prueba['SkinThickness'] <= 65]
df_prueba = df_prueba[df_prueba['DiabetesPedigreeFunction'] >= 0]
```

```python
# Veamos el resultado de eliminar los datos atípicos
fig, axes = plt.subplots(1,2)

sns.boxplot(data=df_prueba['SkinThickness'], ax = axes[0]).set_title('Sin Valores Atípicos')
sns.boxplot(data=df_prueba['Pregnancies'], ax = axes[1]).set_title('Boxplot Normal')

plt.show()
```
